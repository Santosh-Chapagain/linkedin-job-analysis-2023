```python
df.drop_duplicates(inplace=True)
```

```python
jobs = df["Designation"]
```

```python
from collections import Counter
freq = Counter(jobs)

dict1 = freq.items()
```

```python
dict1
```

```python
df_job = pd.DataFrame(dict1 , columns=["Jobs" , "Number_of_Jobs"])
```

```python
df_job.sort_values("Number_of_Jobs" , ascending=False ,ignore_index=True , inplace=True)
```

```python
df_job
```

```python
top_10_job = df_job.loc[0:9]
```

```python
top_10_job
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x= "Jobs" , y="Number_of_Jobs" , data=top_10_job)
plt.xticks(rotation=45, ha='right')
```

Most demanding job is "Team Lead" in the market.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(12, 6))
sns.barplot(x= "Company_Names" , y= "Hiring" , data=df_hired)
plt.xticks(rotation=45, ha='right')
```

Above barplot shows which company hire more applicants .

```python

skill_columns = ['REACTJS', 'AI', 'UI', 'TABLEAU' , 'NODEJS', 'EXCEL', 'POWER BI', 'SELENIUM', 'HTML', 'ML']

top_skills = df[skill_columns].sum().sort_values(ascending=False)
```

```python
top_skills
```

```python
plt.figure(figsize=(12 , 6))
top_skills.plot(kind='bar', title="Top In-Demand Skills", color='skyblue')
plt.ylabel("Number of Job Posts")
plt.tight_layout()
plt.xticks(rotation = 0)
plt.show()
```

As per graph "AI" is the most demanding skill in market .

```python
for skill in skill_columns:
    df_skill = df[df[skill] == 1]
    top_locations = df_skill['Location'].value_counts().head(5)

    print(f"\nTop Locations Hiring for {skill}:")
    print(top_locations)
```

```python
import matplotlib.pyplot as plt

skill_columns = ['AI', 'ML', 'REACTJS', 'HTML','TABLEAU', 'NODEJS', 'EXCEL', 'POWER BI', 'SELENIUM']
n_skills = len(skill_columns)


fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12))
axes = axes.flatten()

for idx, skill in enumerate(skill_columns):
    df_skill = df[df[skill] == 1]
    top_locs = df_skill['Location'].value_counts().head(5)

    axes[idx].barh(top_locs.index[::-1], top_locs.values[::-1], color='orange')
    axes[idx].set_title(f"{skill} - Top Locations")
    axes[idx].set_xlabel("Job Count")
    axes[idx].set_ylabel("Location")

plt.tight_layout()
plt.suptitle("Top Hiring Locations per skill (2023 India)", fontsize=18, y=1.02)
plt.show()
```
